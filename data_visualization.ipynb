```python
import pandas as pd
import io
```

```python
a = pd.read_csv(io.StringIO(uploaded['diaDiem.csv'].decode('utf-8')))
```

```python
a
```

```python
import folium
```

```python
map = folium.Map(location=[10.777400127642721, 106.69536811139102], zoom_start = 12)
```

```python
b = folium.map.FeatureGroup()
```

```python
for lat, lng, des in zip(a.Lat, a.Lng, a.Description):
    b.add_child(
        folium.Marker(
            [lat, lng],
            popup = des
            
))
b.add_to(map)
```

```python
map2 = folium.Map(location=[10.777400127642721, 106.69536811139102], zoom_start = 12)
c = 'Lê Thánh Tôn'
e = 'Lê Đức Thọ'
d = folium.map.FeatureGroup()
for lat, lng, label, des in zip(a.Lat, a.Lng, a.streetName, a.Description):
  if c == label or e == label:
    d.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6,
            popup = '<h1>' + des + '</h1>'
))
d.add_to(map2)
map2
```
